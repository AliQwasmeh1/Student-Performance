#Does age significantly influence academic performance?

```python
plt.Figure(figsize=(5,5))
sns.scatterplot(data=df,x='age',y='overall_score')
plt.title('Distripution by Age and academic performance ')
plt.xlabel='age'
plt.ylabel='overall_score'
plt.show()
```

There is no effect

Does gender and school type influence student performance?

```python
score_means = df.groupby(['gender','school_type'])['overall_score'].mean().unstack()
sns.heatmap(score_means, annot=True, cmap='coolwarm')
plt.title('Average Overall Score by Gender and School Type')
plt.show()
```

Does parents_education level influence academic performance?

```python
plt.Figure(figsize=(5,5))
sns.barplot(data=df,x='parent_education',y='overall_score')
plt.title('distipution by parent_education	and overall_score')
plt.show()
```

Does internet access and participation in extracurricular activities affect performance?

```python
df.groupby(['internet_access', 'extra_activities'])[['math_score', 'science_score', 'english_score', 'overall_score']].mean()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='internet_access', y='overall_score', hue='extra_activities', data=df, errorbar=None)
plt.title('Overall Score by Internet Access and Extra Activities')
plt.show()
```

Which study method leads to the best academic performance?

```python
plt.Figure(figsize=(5,5))
sns.barplot(data=df,x='study_method',y='overall_score')
plt.xlabel='study_method'
plt.ylabel='overall_score'
plt.show()
```
